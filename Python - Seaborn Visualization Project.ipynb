## Project: Nutrition Data Visualization on 80 cereal products
Link: https://www.kaggle.com/datasets/crawford/80-cereals

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
cereal_df = pd.read_csv(r'C:\Users\taran\OneDrive\Desktop\Seaborn\Data\cereal.csv')
```

```python
cereal_df.head()
```

```python
#Irrelevant fields

fields = ['shelf', 'weight', 'cups', 'rating']
cereal_df_new = cereal_df.drop(fields, axis=1)
cereal_df_new
```

```python
cereal_df_new['name'] = cereal_df_new ['name'].apply(str)
```

```python
# Drop non-numeric columns
cereal_numeric = cereal_df_new.select_dtypes(include='number')
cereal_numeric
```

```python
# Calculate correlation
cereal_corr = cereal_numeric.corr()
cereal_corr
```

```python
# ones_like can build a matrix of booleans (True, False) with the same shape as our data
ones_corr = np.ones_like(cereal_corr, dtype=bool)
ones_corr
```

```python
ones_corr.shape, cereal_corr.shape
```

```python
# np's triu: return only upper triangle matrix
mask = np.triu(ones_corr)
mask
```

```python
sns.heatmap(data=cereal_corr, mask=mask)
```

```python
# When removing the upper tri-angle, there are still 2 empty cells in our matris (calories & vitamin)

adjusted_mask = mask [1:, :-1]
adjusted_mask 
```

```python
adjusted_cereal_corr = cereal_corr.iloc[1:, :-1]
```

```python
fig, ax = plt.subplots(figsize=(10,8))
sns.heatmap(data=adjusted_cereal_corr, mask=adjusted_mask,
           annot=True, annot_kws={"fontsize":13},fmt=".2f", cmap=['Pink','#7fc97f'],
           vmin=-1,vmax=1, linecolor='white',linewidths=0.5);

yticks = [i.upper()for i in adjusted_cereal_corr.index]
xticks = [i.upper()for i in adjusted_cereal_corr.columns]

ax.set_yticklabels(yticks,rotation=0, fontsize=13)
ax.set_xticklabels(xticks,rotation=90, fontsize=13)
title = 'CORRELATION MATRIX\nSAMPLED CEREAL COMPOSITION\n'
ax.set_title(title, loc='left', fontsize=18)
```
